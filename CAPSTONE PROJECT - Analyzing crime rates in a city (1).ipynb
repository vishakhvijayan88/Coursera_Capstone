# Analyzis of crimes in City of Chicago

### Introduction
    
    This project is part of the IBM Data Science Professional Certification Capstone Project. This project is for analyzing the crimes in Chicago city and propose remedial measures to curb the crime rate for the 
    respective stake holders. City of Chicago crime rate is quite high among other cities in Illinois. The city has a wide network of surveilance camera providing visuals of every nook and corner of the city. They started installing of cameras in 2003 but still the crimes haven't stopped in the city. This report will help to distinguish vulnerable and safe neighborhoods, which will help to decide on increasing or decreasing security and patroling in those areas.

### Datasets

    This project uses the crime data from City of Chicago Portal which uses data from 'Citizen Law Enforcement Analysis and Reporting'. Further for location vise visualization and analysis, FourSquare API will be used. Also, to analyze the number of officers to population, "Governing calculations of employment and population data from 2015 FBI Uniform Crime Reporting program" data will be used.
