# Analyzis of crimes in City of Chicago

### Introduction
    
    This project is part of the IBM Data Science Professional Certification Capstone Project. This project is for analyzing the crimes in Chicago city and propose remedial measures to curb the crime rate for the 
    respective stake holders. 

### Datasets

    This project uses the crime data from City of Chicago Portal which uses data from 'Citizen Law Enforcement Analysis and Reporting'. Further for location vise visualization and analysis, FourSquare API will be used.
